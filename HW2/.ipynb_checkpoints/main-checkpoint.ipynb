### Data Processing

```python
df = pd.read_excel('2014-2020.xlsx')
```

```python
columns = df.columns.to_list()
columns[0] = 'time'
df.columns = columns
```

```python
df['hour'] = [int(i.split('-')[0]) for i in df['time']]
df['date'] = [i.split('-')[1] for i in df['time']]
df['day'] = [int(i.split('.')[0]) for i in df['date']]
df['month'] = [int(i.split('.')[1]) for i in df['date']]
df['year'] = [int(i.split('.')[2]) for i in df['date']]
```

```python
df = df.drop(columns=['Unnamed: 12', 'date', 'time'])
```

```python
df
```

```python
df_1 = df.groupby('year').sum().iloc[:,:6].reset_index()
```

```python
df_2 = df.copy()
df_2['day_of_year'] = [datetime(y, m, d).timetuple().tm_yday for d, m, y in df_2.iloc[:, -3:].values]
df_2 = df_2.groupby(['hour', 'day_of_year']).mean()['CONSUMPTION'].reset_index()
df_2_1 = df_2.groupby(['day_of_year']).mean()['CONSUMPTION'].reset_index()
```

```python
df_3 = df.groupby('hour').mean().iloc[:,:6].reset_index()
```

```python
df_4 = df.groupby(['hour', 'month']).mean().reset_index()[['hour', 'month', 'CONSUMPTION']]
months = np.array([0, 'January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December'])

def season(month):
    ind = months.tolist().index(month)
    if ind < 3 or ind == 12:
        return "Winter"
    if ind >=3 and ind < 6:
        return "Spring"
    if ind >=6 and ind < 9:
        return "Summer"
    else:
        return "Autumn"

def season_month(month):
    ind = months.tolist().index(month)
    return (ind)%3

df_4['month'] = months[df_4['month']]
df_4['season'] = [season(month) for month in df_4['month']]
df_4['season_month'] = [season_month(month) for month in df_4['month']]
```

```python
df_5 = df.copy()
df_5['weekind'] = [datetime(y, m, d).weekday() for d, m, y in df_5.iloc[:, -3:].values]
df_5['weekday'] = [calendar.day_name[i] for i in df_5['weekind'] ]
df_5 = df_5.groupby('weekday').mean()[['CONSUMPTION', 'weekind']].reset_index().sort_values(by=['weekind'])
df_5['CONSUMPTION'] = [round(i) for i in df_5['CONSUMPTION']]
```

### 1. Power Generation Structure of Ukraine by years

```python
import plotly.express as px

fig = px.area(df_1, x="year", y=['AES', 'TES', 'TEC', 'GES', 'VDE', 'GAES_GEN'], labels={"year": "Year", "variable": "Power Source", 'value': "Generation"}, template="plotly_dark")

fig.update_traces(textposition='top center', mode='lines+markers')

fig.update_layout(
    title = "Annual Generated Power by Source Type, MW",
    autosize=False,
    width=950,
    height=500,
    margin=dict(l=20, r=20, t=40, b=20),
    
    yaxis=dict(
        title_text="Power Generated",
        tickvals=[25e6, 50e6, 75e6, 100e6, 150e6, 125e6, 175e6],
        tickmode="array",
    ),
    
    legend=dict(
        traceorder="reversed",
    )
)

fig.update_yaxes(automargin=True, zeroline=False)
fig.update_xaxes(title_text='Year', zeroline=False)


fig.update_xaxes(showspikes=True, spikethickness=1, spikesnap="cursor", spikemode="across")
fig.update_yaxes(showspikes=True, spikethickness=1)

fig.show()
```

### 2. Power Consumption by Day of the year and Hour

```python
fig = px.scatter(df_2, x="day_of_year", y="hour", size="CONSUMPTION", color="CONSUMPTION", text = 'hour', title="Long-Form Input", labels={"day_of_year": "Day of Year", 'CONSUMPTION': "Consumption, MW", 'hour': "Hour"}, template="plotly_dark")
fig.update_layout(
    title = "Daily Average Power Consumption by Day of the Year, MW",
    autosize=False,
    width=950,
    height=500,
    margin=dict(l=20, r=20, t=40, b=20),
  
    yaxis=dict(
        title_text="Hour",
        tickvals=list(range(1,25)),
        tickmode="array",
    ),
    
    legend=dict(
    )
)

fig.update_yaxes(automargin=True, zeroline=False)
fig.update_xaxes(title_text='Day of Year', zeroline=False)


fig.update_xaxes()
fig.update_yaxes()

fig.update_traces(marker_line_color='rgb(8,48,107)',
                  marker_symbol=23,
                  marker_line_width=0.0, opacity=0.3)

fig.show()
```

### 3. Power Generation Structure of Ukraine by hours

```python
import plotly.express as px

fig = px.bar(df_3, x="hour", y=['AES', 'TES', 'TEC', 'GES', 'VDE', 'GAES_GEN'], labels={"hour": "Hour", "variable": "Power Source", 'value': "Generation"}, template="plotly_dark")

fig.update_layout(
    title = "Daily Average Power Generation Structure, MW",
    autosize=False,
    width=950,
    height=500,
    margin=dict(l=20, r=20, t=40, b=20),
    yaxis=dict(
        title_text="Electricity Generation",
        tickmode="array",
    ),
    xaxis=dict(
        title_text="Consumption",
        tickvals=list(range(1, 25)),
    ),
    
    legend=dict(
        traceorder="reversed"
    )
)

fig.update_yaxes(automargin=True, zeroline=False)
fig.update_xaxes(title_text='Hour', zeroline=False)


fig.update_xaxes(showspikes=True, spikethickness=1, spikesnap="cursor", spikemode="across")
fig.update_yaxes(showspikes=True, spikethickness=1)

fig.show()
```
